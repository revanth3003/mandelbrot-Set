```python
import numpy
import warnings
import matplotlib.pyplot as plt
```

```python
#grid of points(x,y) 500 × 500 grid
x,y = numpy.mgrid[-2:1:500j, -1.5:1.5:500j]
c = x+1j*y

z = c

#threshold
N_max = 50 

warnings.filterwarnings('ignore')

#zn+1 =z^2n +c
for j in range(N_max):
    z = z**2 + c
    mask = (abs(z) < 50)

#plot the fractal and save it as an image
plt.imshow(mask.T, extent=[-2, 1, -1.5, 1.5])
plt.gray()
plt.savefig('mandelbrot.png')
plt.show()
```
